```python
# 1. IMPORTACIONES Y CONFIGURACIÓN

import pandas as pd
import numpy as np
import plotly.express as px
import unicodedata
import warnings

warnings.filterwarnings('ignore')

print("Librerías importadas correctamente")
```

```python
# 2. CARGA DE LOS DATOS
df = pd.read_csv("Precios_Combustibles_(2017-1)_20260420.csv", encoding='utf-8')
print("Archivo cargado correctamente")
print(f"Dimensiones del dataset: {df.shape[0]:,} filas × {df.shape[1]} columnas")
```

```python
# 3. EXPLORACIÓN INICIAL

print("=== DIMENSIONES DEL DATAFRAME ===")
print(df.shape)

print("\n=== INFORMACIÓN GENERAL ===")
df.info()

print("\n=== PRIMERAS 20 FILAS ===")
display(df.head(20))

print("\n=== ÚLTIMAS 20 FILAS ===")
display(df.tail(20))

print("\n=== VALORES ÚNICOS EN COLUMNAS IMPORTANTES ===")
print("Productos:", df["producto"].unique())
print("\nMarcas (Bandera):", df["bandera"].unique())
print("\nDepartamentos:", df["departamentonombre"].unique()[:15])

print("\n=== ESTADÍSTICAS DEL PRECIO ===")
print(df["precio"].describe())
```

```python
# 4. ELIMINACIÓN Y RENOMBRADO DE COLUMNAS
# Eliminar columna que no aporta información
data = df.drop(columns=['estado'])
data.drop_duplicates(inplace=True)
# Renombrar columnas 
data.rename(columns={'departamentocodigo': 'depto_codigo'}, inplace=True)
data.rename(columns={'departamentonombre': 'departamento'}, inplace=True)
data.rename(columns={'municipiocodigo': 'mun_codigo'}, inplace=True)
data.rename(columns={'municipionombre': 'municipio'}, inplace=True)
data.rename(columns={'fecharegistro': 'fecha_registro'}, inplace=True)





print("Columnas renombradas correctamente")
print("Columnas actuales:", data.columns.tolist())

```

```python
# 1. Multiplicamos por 1000 para pasar de decimales a miles
data['precio'] = data['precio'] * 1000

# 2. Lo convertimos a entero (int) para quitar el ".0" y que se vea limpio
data['precio'] = data['precio'].astype(int)

# 3. Vemos el resultado
data.head()
```

```python
# 5. NORMALIZACIÓN DE TEXTO

def normalize_text(text):
    if isinstance(text, str):
        # Quitar acentos y convertir a mayúsculas
        text = unicodedata.normalize('NFKD', text).encode('ASCII', 'ignore').decode('ASCII')
        return text.upper().strip()
    return text

# Aplicar normalización a las columnas de texto
cols_to_normalize = ['departamento', 'municipio', 'agente', 'bandera', 'producto']
for col in cols_to_normalize:
    data[col] = data[col].apply(normalize_text)

print("Texto normalizado (sin acentos y en mayúsculas)")
```

```python
# 6. CORRECCIÓN DE FECHAS Y NUEVAS COLUMNAS
# Convertir la columna de fecha a formato datetime
data['fecha'] = pd.to_datetime(data['fecha_registro'],format='%Y %b %d %I:%M:%S %p',errors='coerce')

# Crear columnas útiles para análisis
data['año'] = data['fecha'].dt.year
data['mes'] = data['fecha'].dt.month
data['nombre_mes'] = data['fecha'].dt.strftime('%B')

# Simplificar el tipo de producto
def simplificar_producto(p):
    if 'GASOLINA CORRIENTE' in str(p):
        return 'GASOLINA CORRIENTE'
    elif 'GASOLINA EXTRA' in str(p):
        return 'GASOLINA EXTRA'
    elif any(x in str(p) for x in ['BIODIESEL', 'ACPM', 'DIESEL']):
        return 'DIESEL / BIODIESEL'
    elif 'KEROSENE' in str(p):
        return 'KEROSENE'
    else:
        return p

data['producto_simplificado'] = data['producto'].apply(simplificar_producto)

data.drop(columns=['fecha_registro', 'mes'], inplace=True)

print("Fechas corregidas y columnas nuevas creadas")
print(data['producto_simplificado'].value_counts())



```

```python
#Traduce el nombre del mes 

data['nombre_mes'] = data['fecha'].dt.strftime('%B')


meses_espanol = {'January': 'Enero', 'February': 'Febrero', 'March': 'Marzo', 'April': 'Abril', 'May': 'Mayo', 'June': 'Junio', 'July': 'Julio', 'August': 'Agosto', 'September': 'Septiembre', 'October': 'Octubre', 'November': 'Noviembre', 'December': 'Diciembre'}
data['nombre_mes'] = data['nombre_mes'].map(meses_espanol)

# 3. Ahora sí, elimina lo que no sirve
# Verifica que 'mes' y 'fecha_registro' existan antes de borrarlas
data.drop(columns=['fecha_registro', 'mes'], errors='ignore', inplace=True)
```

```python
data.head()
```

```python
# =============================================
# 7. DETECCIÓN DE OUTLIERS EN PRECIOS
# =============================================

Q1 = data['precio'].quantile(0.25)
Q3 = data['precio'].quantile(0.75)
IQR = Q3 - Q1

outliers = data[(data['precio'] < (Q1 - 1.5 * IQR)) | (data['precio'] > (Q3 + 1.5 * IQR))]

print(f"Total de outliers detectados: {len(outliers)}")
if len(outliers) > 0:
    print("\nEjemplos de outliers:")
    display(outliers[['departamento', 'municipio', 'producto_simplificado', 'precio', 'bandera']].head(10))
```

```python
# 8. EXPORTACIÓN DEL ARCHIVO LIMPIO


data.to_excel("Precios_Combustibles_LIMPIO.xlsx", index=False)
print("Archivo limpio guardado como: Precios_Combustibles_LIMPIO.xlsx")
```

```python
# 9. GRÁFICOS Y ANÁLISIS VISUAL

# Comparación de Marcas
marcas_principales = ['TERPEL', 'BIOMAX', 'MOBIL', 'TEXACO', 'PUMA']
df_marcas = data[data['bandera'].isin(marcas_principales)]
fig2 = px.box(df_marcas, x='bandera', y='precio', color='producto_simplificado',
              title="Comparación de Precios por Marca")
fig2.show()



# Distribución por tipo de combustible
fig5 = px.histogram(data, x="producto_simplificado", color="producto_simplificado",title="Cantidad de Registros por Tipo de Combustible")
fig5.show()
```

```python
# Evolución temporal del precio
# 1. Agrupamos por Mes y Tipo de Producto para limpiar la tendencia
precio_final = data.set_index('fecha').groupby([pd.Grouper(freq='M'), 'producto_simplificado'])['precio'].mean().reset_index()

# 2. Creamos la gráfica comparativa
# Una línea por cada combustible
fig4 = px.line(precio_final, x='fecha',y='precio',color='producto_simplificado', title="Comparativa Mensual: Evolución de Precios por Combustible (2015-2018)",markers=True)

# 3. Ajustamos el diseño para que sea profesional
fig4.update_layout(xaxis_title="Periodo (Mes/Año)",yaxis_title="Precio Promedio (COP)",legend_title="Tipo de Combustible",hovermode="x unified" # Para ver todos los precios al pasar el mouse por una fecha
)
```

```python

# GRÁFICO: Precio Promedio por Marca Y Tipo de Combustible


# Calcular precio promedio por Marca y por Tipo de Combustible
precio_marca_tipo = data.groupby(['bandera', 'producto_simplificado'])['precio'].mean().reset_index()

# Ordenar por precio promedio descendente
precio_marca_tipo = precio_marca_tipo.sort_values(by='precio', ascending=False)

fig8 = px.bar(precio_marca_tipo, x='bandera', y='precio',color='producto_simplificado',      title="Precio Promedio por Marca y Tipo de Combustible",labels={'precio': 'Precio Promedio (COP)', 'bandera': 'Marca','producto_simplificado': 'Tipo de Combustible'},text_auto='.2f',barmode='group')   # Barras agrupadas

fig8.update_layout(xaxis_tickangle=-45, height=700,legend_title="Tipo de Combustible")

fig8.show()

print("Gráfico: Precio por Marca y Tipo de Combustible")
```

```python
# GRÁFICO: Top 15 Municipios más caros


precio_municipio = data.groupby('municipio')['precio'].mean()\
                    .sort_values(ascending=False).head(15).reset_index()

fig10 = px.bar(precio_municipio, x='municipio',  y='precio',title="Top 15 Municipios con los Precios de Combustible más Altos",labels={'precio': 'Precio Promedio (COP)', 'municipio': 'Municipio'}, color='precio',text_auto='.2f')

fig10.update_layout(xaxis_tickangle=-45, height=650)
fig10.show()

print("Top 15 Municipios más caros")
```

```python

# GRÁFICO: Cantidad de Estaciones por Departamento


estaciones_depto = data.groupby('departamento').size()\
                    .sort_values(ascending=False).head(12).reset_index()
estaciones_depto.columns = ['departamento', 'cantidad_estaciones']

fig12 = px.bar(estaciones_depto, x='departamento', y='cantidad_estaciones',title="Top 12 Departamentos con Más Estaciones de Servicio",labels={'cantidad_estaciones': 'Número de Estaciones', 'departamento': 'Departamento'},color='cantidad_estaciones',text_auto=True)
fig12.update_layout(xaxis_tickangle=-45, height=600)
fig12.show()

print("Gráfico de Estaciones por Departamento")
```
